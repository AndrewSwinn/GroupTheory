```python
def rotate(cubie, direction):
    num, dir = cubie & 0x0F, cubie & 0xF0
    
    if   direction == 'A':
        if   dir==0x00: dir=0x80
        elif dir==0x40: dir=0x00
        elif dir==0x80: dir=0x40
    elif direction == 'C':
        if   dir==0x00: dir=0x40
        elif dir==0x40: dir=0x80
        elif dir==0x80: dir=0x00
    
    return num ^ dir

edges   = [0x01, 0x02, 0x03, 0x04, 0x05, 0x06, 0x07, 0x08, 0x09, 0x0A, 0x0B, 0x0C]
corners = [0x01, 0x02, 0x03, 0x04, 0x05, 0x06, 0x07, 0x08]
cube    = [edges, corners]
members = {' ': cube}
```
